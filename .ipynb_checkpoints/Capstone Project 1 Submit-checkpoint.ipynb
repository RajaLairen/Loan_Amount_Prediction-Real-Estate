```python
def error_ploter(data):
    """
    This function plot null value contain % for each features
    param:
        data - the dataset to which you want to plot the null value % for each features
    """
    null_percentage=data.isna().sum().sort_values(ascending=False)/data.shape[0]*100
    plt.figure(figsize=(10,10))
    plot=sns.barplot(x=null_percentage.values,y=null_percentage.index)
    plt.title('Null value percentage for individual features',size=18,fontweight='bold')
    plt.xlabel('Null value percentage',size=15)
    plt.ylabel('Features name',size=15)

    for rec in plot.patches:
        val=rec.get_width()
        height=rec.get_height()
        y=rec.get_y()
        plot.text(val+1,y+1,f'{round(val,2)} %')
    save_plots('Null value percentage for individual features')
    plt.show() 
    return null_percentage
```

```python
def imputer(data,err):
    '''
    A function to impute a dataframe. The strategy was to impute the categorical features with most_frequent and the numerical continuous column with median
    params:
        data: the dataset to impute
        err: the null values rate of the the given dataset
    '''
    # Get the features that have the null rate greater than 85%
    max_err_index=err[err>=85].index
    # Get the features that have null values greater that 0 and less that 85
    null_index=err[(err>0) & (err<85)].index
    
    imputer_median=SimpleImputer(strategy='median')
    imputer_mode=SimpleImputer(strategy='most_frequent')
    for col in null_index:
        if(data[col].dtypes!='O'):
            data[col]=imputer_median.fit_transform(data[[col]])
        else:
            print(col)
            data[col]=imputer_mode.fit_transform(data[[col]])
        
        
    # Drop the featues that have null rate greater than 85%
    data=data.drop(max_err_index,axis=1)
    return data
```
